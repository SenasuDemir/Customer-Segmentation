# <font color='#038C3E'> CUSTOMER SEGMENTATION

## <font color='#038C3E'> Project Overview

The goal of this project is to segment a company's customer base to enhance targeted marketing strategies and gain insights into customer buying behaviors. By leveraging customer, product, and order data, we aim to create a comprehensive segmentation model that groups customers based on their purchasing patterns. This segmentation will provide valuable insights into how different customer groups interact with the company's products and services. The project will utilize KMeans clustering to identify distinct customer segments and will include the creation of an RFM table to measure customer behavior based on Recency, Frequency, and Monetary metrics. The RFM analysis will allow for the classification of customers into different groups based on their recent activity, frequency of purchases, and total spending. Additionally, the optimal number of segments will be determined, and the characteristics of each segment will be analyzed to guide marketing decisions.

## <font color='#038C3E'> RFM (Recency-Frewuency-Monetary)
- **Recency:** the date they made their last purchase
- **Frequency:** how often they make purchases
- **Monetary:**  the amount spent on purchases

```python
from datetime import datetime
data=pd.DataFrame()
data['PurchaseDate']=pd.to_datetime(df['Orders.placed_date'])

# Recency Calculation
data['Recency'] = (datetime.now() - data['PurchaseDate']).dt.days

# Frequency Calculation
data['CustomerID']=df['Customers.id']
data['OrderID']=df['Orders.id']
frequency_data=data.groupby('CustomerID')['OrderID'].count().reset_index()
frequency_data.rename(columns={'OrderID':'Frequency'},inplace=True)
data=data.merge(frequency_data,on='CustomerID',how='left')

# Monetary Calculation
data['TransactionAmount']=df['Orders.total']
monetory_data=data.groupby('CustomerID')['TransactionAmount'].sum().reset_index()
monetory_data.rename(columns={'TransactionAmount':'MonetaryValue'},inplace=True)
data=data.merge(monetory_data,on='CustomerID',how='left')
data.head()
```

```python
recency_scores=[5,4,3,2,1] #higher score for lower recency
frequency_scores=[5,4,3,2,1] # higher score for higher frequency
monetary_scores=[5,4,3,2,1] # higher score for higher monetary value

data['RecencyScore']=pd.cut(data['Recency'],bins=5,labels=recency_scores)
data['FrequencyScore']=pd.cut(data['Frequency'],bins=5,labels=frequency_scores)
data['MonetaryScore']=pd.cut(data['MonetaryValue'],bins=5,labels=monetary_scores)

data['RecencyScore']=data['RecencyScore'].astype(int)
data['FrequencyScore']=data['FrequencyScore'].astype(int)
data['MonetaryScore']=data['MonetaryScore'].astype(int)
```

```python
data['RFM_Score']=data['RecencyScore']+data['FrequencyScore']+data['MonetaryScore']
```

```python
data.head()
```

```python
data['RFM_Score'].value_counts()
```

```python
bins = [0, 9, 12, 15]  
labels = ['Low', 'Mid', 'High']
data['Value Segment'] = pd.cut(data['RFM_Score'], bins=bins, labels=labels, right=False)
```

```python
data.sample(5)
```

```python
# RFM Segment Distribution
segment_counts = data['Value Segment'].value_counts().reset_index()
segment_counts.columns = ['Value Segment', 'Count']

fig_segment_dist=px.bar(segment_counts,
                        x='Value Segment',
                        y='Count',
                        color='Value Segment',
                        color_discrete_sequence=px.colors.qualitative.Pastel,
                        title='RFM Value Segment Distribution'
                       )
fig_segment_dist.update_layout(xaxis_title='RFM Value Segment',
                              yaxis_title='Count',
                              showlegend=False)
```

```python
data['RFM Customer Segments']=''

data.loc[data['RFM_Score']>=9,'RFM Customer Segments']='Champions'
data.loc[(data['RFM_Score']>=6) & (data['RFM_Score']<9),'RFM Customer Segments']='Potential Loyalists'
data.loc[(data['RFM_Score']>=5) & (data['RFM_Score']<6),'RFM Customer Segments']='At Risk Customers'
data['RFM Customer Segments'].value_counts()
```

```python
# Distribution of customers across different RFM customer segments
segment_product_counts=data.groupby(['Value Segment','RFM Customer Segments']).size().reset_index(name='Count')
segment_product_counts=segment_product_counts.sort_values('Count',ascending=False)
fig_treemap_segment_product=px.treemap(segment_product_counts,
                                      path=['Value Segment','RFM Customer Segments'],
                                      values='Count',
                                      color='Value Segment',
                                      color_discrete_sequence=px.colors.qualitative.Pastel,
                                       title='RFM Customer Segments by Value'
                                      )
fig_treemap_segment_product.show()
```

## <font color='#038C3E'> Results

The RFM (Recency, Frequency, and Monetary) analysis was conducted to segment customers based on their purchasing behavior. The analysis categorized customers into different segments based on their recency of purchase, frequency of transactions, and monetary value spent. The RFM score was calculated for each customer, with higher scores indicating more valuable customers.

From the results, customers who exhibit high Recency, Frequency, and Monetary scores (with an RFM score of 13) are classified as "High" value customers. These customers are highly engaged, purchase frequently, and contribute significantly to the company's revenue. On the other hand, customers with lower scores, such as an RFM score of 5, are identified as "Low" value customers. These customers have either infrequent purchases, low transaction amounts, or both, and may require more targeted marketing efforts to increase their engagement.

The silhouette score for the clustering model was calculated to be 0.587, suggesting a moderate to strong cluster separation, which supports the effectiveness of the segmentation model in distinguishing customer groups based on their purchasing behavior.
